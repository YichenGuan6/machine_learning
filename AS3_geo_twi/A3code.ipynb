# Read CSV file

```python
import pandas as pd
import re

#####training data
tfidf_train_raw = pd.read_csv("train_tfidf.csv")
tfidf_train = tfidf_train_raw["tweet"]
####development data
tfidf_dev_raw = pd.read_csv("dev_tfidf.csv")
tfidf_dev = tfidf_dev_raw["tweet"]
```

# Get Label

```python
train_label = tfidf_train_raw["region"]
dev_label = tfidf_dev_raw["region"]
# print(train_label.value_counts())
# print(dev_label.value_counts())
```

# read vocab.txt file and save in dictionary

```python
vocab = open("vocab.txt",'r',errors='replace')
dic_vocab = {}
for i in vocab.readlines():
    value,key = i.replace("\n","").split("\t")
    key = int(key)
    dic_vocab[key] = value
```

# Refomulate the data set structure

```python
def ref_data_set(data):
    result = []
    for i in data:
        temp = re.findall(r'[(](.*?)[)]',i)
        new_line = [0 for i in range(len(dic_vocab))]
        for j in temp:
            index,value = j.split(",")
            new_line[int(index)] = float(value)
        result.append(new_line)
    return result  
```

```python
new_train_tfidf = ref_data_set(tfidf_train)
new_train_tfidf_frame = pd.DataFrame(new_train_tfidf)
#new_train_tfidf_frame.columns = list(dic_vocab.values())
new_train_tfidf_frame
```

```python
new_dev_tfidf = ref_data_set(tfidf_dev)
new_dev_tfidf_frame = pd.DataFrame(new_dev_tfidf)
#new_dev_tfidf_frame.columns = list(dic_vocab.values())
new_dev_tfidf_frame
```

# Define two type Naive Bayes models

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import CategoricalNB

from sklearn.naive_bayes import ComplementNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import heapq

def Mul_NB (train_features, train_labels, test_features):
    predictions = []
    multinomial_bn = MultinomialNB()
    multinomial_bn.fit(train_features,train_labels)
    predictions = multinomial_bn.predict(test_features)
    return predictions

def Ber_NB (train_features, train_labels, test_features):
    predictions = []
    BernNB = BernoulliNB(alpha=1.0)
    BernNB.fit(train_features,train_labels)
    predictions = BernNB.predict(test_features)
    return predictions
```

# define two type logistics regression models

```python
def LR_class(train_features, train_labels, test_features):
    predictions = []
    lr = LogisticRegression(max_iter = 1000)
    lr.fit(train_features,train_labels)
    predictions = lr.predict(test_features)
    return predictions
```

```python
def ovo_LR_class(train_features, train_labels, test_features):
    predictions = []
    lr = LogisticRegression(multi_class = "multinomial",solver = "newton-cg",max_iter = 1500)
    lr.fit(train_features,train_labels)
    predictions = lr.predict(test_features)
    return predictions
```

# Stratified random sampling

```python
data_new =pd.concat([new_train_tfidf_frame,train_label],axis=1)
data_new
```

```python
def typicalsamling(group,typicalNDict):
    name=group.name
    n=typicalNDict[name]
    return group.sample(n=n)
```

```python
import numpy
numpy.random.seed(seed=2)
typicalNDict={
        'NORTHEAST':7000,
        'SOUTH':7000,
        'WEST':7000,
        'MIDWEST':7000
        }

result=data_new.groupby(
        'region',group_keys=False
        ).apply(typicalsamling,typicalNDict)
```

```python
sRs_train_label = result["region"]
sRs_train_tfidf = result.drop(["region"],axis=1)
sRs_train_tfidf
```
